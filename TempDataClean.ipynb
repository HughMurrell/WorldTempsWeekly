# Cleaning average monthly temperature data from NOAA

Two files were generated from NOAA data stored in:

ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/v4/ghcnm.tavg.latest.qcf.tar.gz

stations.csv is a csv version of ghcnm.tavg.v4.0.1.20200305.qcf.inv

temps.csv is a csv version of ghcnm.tavg.v4.0.1.20200305.qcf.dat

The csv files were created from these fixed column width data sets 
using unix tools such as sed and vi. Field headers were created
manually. Records with missing values were removed.

This notebook shows how to read in both csv files and uses
the station code for each station to pick out the latest full
temperature record for each station. 

The records are then combined and written out as one csv file
called clean_data.csv. 

only clean_data.csv resides in the github repository as the
original data sets are too large

```julia
# first make sure the `CSV` package is loaded
using CSV
```

### Load the station names and lat/long data

```julia
# code,lat,long,height,name
sdata = CSV.read("data_stations.csv",silencewarnings=true)
```

### Load the monthly  average temperature records for each station

```julia
# code,year,tavg,t1,f1,t2,f2,t3,f3,t4,f4,t5,f5,t6,f6,t7,f7,t8,f8,t9,f9,t10,f10,t11,f11,f12,f12
tdata = CSV.read("data_temps.csv")
```

```julia
tsdata = sort(tdata,[:code,:year],rev=true)
```

### Find the latest full temperature record for each station

```julia
inds = indexin(unique(tsdata[:,:code]), tsdata[:,:code])
latest_temps = tsdata[inds,:]
```

```julia
typeof(latest_temps)
```

```julia
typeof(sdata)
```

```julia
using DataFrames
```

### Join the two dataframes using station code as the key

```julia
# join the two dta sets using the station code field
clean_data = join(sdata, latest_temps, on = :code, kind = :inner)
```

```julia
# drop some unimportant fields
select!(clean_data,Not([:f1,:f2,:f3,:f4,:f5,:f6,:f7,:f8,:f9,:f10,:f11,:f12]))
```

```julia
# have a look at which years most of the data were recorded
using Plots
histogram([ clean_data[i,:year] for i in 1:size(clean_data,1)])
```

### Save the cleaned monthly data for weekly interpolation 

see the TempInterp notebook

```julia
CSV.write("data_clean.csv",clean_data)
```
