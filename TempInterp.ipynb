#### Distance metric for two points on the globe

computes "as the crow flies" great circle distance in km

see Spherical Law of Cosines from:
https://www.movable-type.co.uk/scripts/latlong.html

```julia
dist(lat1,long1,lat2,long2) = 
  acos( sin(lat1*pi/180)*sin(lat2*pi/180) + 
    cos(lat1*pi/180)*cos(lat2*pi/180)*cos(long2*pi/180-long1*pi/180) ) * 6371
```

```julia
dist(1,20,-1,20)
```

```julia
dist(179,20,-179,20)
```

```julia
dist(180,0,0,0)
```

```julia
function closest_station(lat, long,data)
    distances = [ dist(lat,long,data[i,:lat],data[i,:long]) for i in 1:size(data,1) ]
    best_ind = argmin(distances)
    return data[best_ind,:]
end
```

```julia
airports = CSV.read("airports.csv",header = false)
```

```julia
names!(airports, [:id,:name,:district,:country,:code,:country_code,
    :lat,:long,:height,:c10,:c11,:c12,:c13,:c14])
```

```julia
function estimate_weekly_temps(ap_data, st_data )
    wxs = collect(-1:7:356)
    wxs[1]=0
    mxs = (1:30:365)
    coords = [ap_data[1:size(ap_data,1),:lat]  ap_data[1:size(ap_data,1),:long] ]
    df = DataFrame(typeof.(first.(eachcol(st_data))),names(st_data))
    wt = reshape([],0,52)
    for i in 1:size(coords,1)
        if i % 100 == 0
            println(i," of ",size(coords,1))
        end
        ct = closest_station(coords[i,1],coords[i,2],st_data)
        mt = [ct[:t1] , ct[:t2], ct[:t3] , ct[:t4], ct[:t5] , ct[:t6],
              ct[:t7] , ct[:t8], ct[:t9] , ct[:t10], ct[:t11] , ct[:t12], ct[:t1]]
        iw = CubicSplineInterpolation(mxs, mt, extrapolation_bc = Periodic())
        wtemps = round.(Int,iw(-15:7:49*7)) 
        append!(df,DataFrame(closest_station(coords[i,1],coords[i,2],st_data)))
        wt = [wt; wtemps']
    end
    wtdf = DataFrame(wt)
    names!(wtdf, Symbol.(["w$i" for i in 1:52]))
    return(hcat(ap_data,df,wtdf,makeunique=true))
end
```

```julia
adf = estimate_weekly_temps(airports,clean_data)
```

```julia
max_temp = maximum(Matrix{Int}(adf[:,Symbol.(["w$i" for i in 1:52])]))
println("max temp = ",max_temp)
min_temp = minimum(Matrix{Int}(adf[:,Symbol.(["w$i" for i in 1:52])]))
println("min temp = ",min_temp)
```

```julia
names(adf)
```

```julia
# compute the worst distance between airport and nearest station
dist_cf = round.(Int, [ dist(adf[i,:lat],adf[i,:long],adf[i,:lat_1],adf[i,:long_1]) 
                for i in 1:size(adf,1) ])

maximum(dist_cf)
```

```julia
# compute average dist between airport and nearest station
sum(dist_cf)/length(dist_cf)
```

```julia
# add distance to nearest station as an extra column of adf
adf[!,:dist_cf] = dist_cf
names(adf)
```

```julia
# save the augmented airport data frame.
CSV.write("airports_temp_augmented.csv",adf)
```

### Create an animation to see if data looks reasonable

```julia
# plot one frame but scale colors between min and max of ALL frames
# as we go through he 52 frames we will print min amd max temp for that week
using Plots
function world_temp(w,adf)
    max_temp = maximum(Matrix{Int}(adf[:,Symbol.(["w$i" for i in 1:52])]))
    min_temp = minimum(Matrix{Int}(adf[:,Symbol.(["w$i" for i in 1:52])]))
    max_week = maximum(adf[:,Symbol("w$w")])
    print("max temp for week w$w = ",max_week)
    min_week = minimum(adf[:,Symbol("w$w")])
    println(" min temp for week w$w = ",min_week)
    colors = adf[:,Symbol("w$w")] .- min_temp
    colors = colors ./ (max_temp - min_temp)
    p = scatter(adf[:,:long], adf[:,:lat], 
        zcolor = colors,
        m = ColorGradient([:blue,:yellow,:red]),
        legend=false,markerstrokewidth=0,title="week_$w")
    plot(p)
end
```

```julia
# end of december should be cold in Northern hemisphere 
# and warm in Southern hemisphere
world_temp(52,adf)
```

```julia
anim = @animate for i ∈ 1:52
    world_temp(i,adf)
end
```

```julia
# now make an animated gif from all 52 frames. 
# (requires unix tool: ffmpeg to be available from the cmd line)
# to view thw animation open anim_world_temp.gif with google chrome
gif(anim, "anim_world_temp.gif", fps = 5)
```
